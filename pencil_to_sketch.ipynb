TASK 2 - Image to Pencil Sketch with Python

```python
import cv2
from google.colab.patches import cv2_imshow
```

READING MAIN IMAGE

```python
image =cv2.imread("4724804.png")

cv2_imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

INVERTED THE GRAYSCALE IMAGE

```python
inverted_gray_image = cv2.bitwise_not(gray_image)
cv2_imshow(inverted_gray_image)
```

APPLYING GUASSIAN BLUR

```python
inverted_blurry= cv2.GaussianBlur(inverted_gray_image, (21 ,21), sigmaX=0, sigmaY =0)
cv2_imshow(inverted_blurry)
```

MERGING GRAY SCALE AND INVERTED BLURRY IMAGE

```python
def dodgev2(x,y):
  return cv2.divide(x, 255 - y, scale = 256)
```

```python
final_pencil_sketch = dodgev2(gray_image, inverted_blurry)
cv2_imshow(final_pencil_sketch)
```

COMPARING ORIGINAL AND PENCIL SKETCH

```python
cv2_imshow(image)
cv2_imshow(final_pencil_sketch)
```
